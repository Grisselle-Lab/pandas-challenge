```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
purchase_data  = "Resources/purchase_data.csv"


# Read purchasing file and store into pandas data frame
purchase_data_df = pd.read_csv(purchase_data)
purchase_data_df.head()
purchase_data_df
```

## Player Count

* Display the total number of players


```python
# display the sum for purchase id as equal to 303810
total_players = purchase_data_df["Purchase ID"].sum()
total_players
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Calculations 

Number_of_Unique_Items = purchase_data_df["Item ID"].sum()
Number_of_Unique_Items

Average_Price = purchase_data_df["Price"].mean()
Average_Price
    
Number_of_Purchases = purchase_data_df["Purchase ID"].sum()
Number_of_Purchases
    
Total_Revenue = purchase_data_df["Price"].sum()
Total_Revenue

#DataFrame creation
Purchasing_Analysis = pd.DataFrame({"Number of Unique Items": [Number_of_Unique_Items],
                              "Average Price": Average_Price,
                              "Number of Purchases": Number_of_Purchases,
                              "Total_revenue": Total_Revenue})
Purchasing_Analysis



```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# The value counts for Gender series
gender_dem_df = pd.DataFrame(purchase_data_df["Gender"].value_counts())
gender_dem_df

# The percentage calculation for Gender
percentage_of_players = (purchase_data_df["Gender"].value_counts()/total_players)*100
percentage_of_players

gender_dem_df["Percentage of Players"] = percentage_of_players
gender_dem_df["Percentage of Players"] = gender_dem_df["Percentage of Players"].map("{:,.2f}%".format)
gender_dem_df

# Change columns 
org_gender_dem_df = gender_dem_df[["Gender", "Percentage of Players"]]
org_gender_dem_df

# Rename the column 
fin_gender_dem_df = org_gender_dem_df.rename(columns={"Gender":"Total Count"})
fin_gender_dem_df

```


## Purchasing Analysis (Gender)

Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender
Create a summary data frame to hold the results
Optional: give the displayed data cleaner formatting
Display the summary data frame

```python
# Using GroupBy for "Gender" values
gender_grouped_purchased_data_df = purchase_data_df.groupby(["Gender"])

# Gender grouped count
gender_grouped_purchased_data_df["Purchase ID"].count().head(10)

# Get total purchase value by gender
total_purchase_value = gender_grouped_purchased_data_df["Price"].sum()
total_purchase_value.head()
dlr_total_purchase_value = total_purchase_value.map("${:,.2f}".format)
dlr_total_purchase_value.head()

# Average purchase price 
avg_purchase_price = gender_grouped_purchased_data_df["Price"].mean()
avg_purchase_price.head()
dlr_avg_purchase_price = avg_purchase_price.map("${:,.2f}".format)
dlr_avg_purchase_price.head()

# Normalized totals
avg_totals = total_purchase_value/gender_grouped_purchased_data_df["Purchase ID"].count()
dlr_avg_totals = avg_totals.map("${:,.2f}".format)
dlr_avg_totals.head()

# Organize summary gender data
org_gender_purchased_data_df = pd.DataFrame(gender_grouped_purchased_data_df["Purchase ID"].count())
org_gender_purchased_data_df["Average Purchase Price"] = dlr_avg_purchase_price  
org_gender_purchased_data_df["Total Purchase Value"] = dlr_total_purchase_value 
org_gender_purchased_data_df["Avg Total Purchas per Person"] = dlr_avg_totals
org_gender_purchased_data_df

# Summary 
summary_gender_purchased_data_df = org_gender_purchased_data_df.rename(columns={"Purchase ID":"Purchase Count"})
summary_gender_purchased_data_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
purchase_data_df = pd.read_csv(purchase_data)
purchase_data_df.head()

#Cleaning and counting data for empty or null fields 
purchase_data_df = purchase_data_df.dropna(how='all')
purchase_data_df.count()

#Counting intances on Age series
Age_counts = purchase_data_df["Age"].value_counts()
Age_counts.head()

#Creating bins and groups lables for discriminatory parameters and indexing
bins = [0, 10, 15, 19, 23, 27, 31, 35, 39, 100]
group_labels = [ "1 to 10", "11 to 14", "15 to 18","19 to 22","23 to 26","27 to 30", "31 to 34","35 to 38","39 to 60"]

#Setting  Age series for bins and group labels
pd.cut(purchase_data_df["Age"], bins, labels=group_labels).head()
purchase_data_df["Age Group"] = pd.cut(purchase_data_df["Age"], bins, labels=group_labels)

# GroupBy by Age Group created on the previous step
binsgroup = purchase_data_df.groupby("Age Group")

#SN series represent the players and we want to know their count per bin
SNtotal = binsgroup["Age"].count()

#Percentage of SN calculation, the total was optained from previous step after cleaning the data.
percentage = SNtotal / 576 * 100

#Data Frame creation
age_demographics = pd.DataFrame({"Percentage of Players" :percentage,"Total Count" :SNtotal}).round(2)
age_demographics.head()



```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Calculations for Price series group by SN or Player
top_spender_df = purchase_data_df.groupby(["SN"]).agg({"Price" : ["count", "mean", "sum"]})
top_spender_df [("Price","sum")]= top_spender_df[("Price", "mean")].map("${:.2f}".format)

#Sort the total value and display
top_spender_df.sort_values ([("Price", "sum")], ascending  = False).head()


```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Retrieve the Item ID, Item Name, and Item Price columns 
#Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value
Most_popular_df = purchase_data_df.groupby(["Item Name", "Item ID"]).agg({"Price"  : ["count", "mean", "sum"]})

#Sort and display for most Popular Item
Most_popular_df.sort_values ([("Price", "count")], ascending = False).head()
```
